#### Adding some docs to vector-store

```python
from proj.vector_store.chroma_vdbstore import DocumentVectorStore
from proj.api.groq_models import embed_client
import os
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
vector_store = DocumentVectorStore(
    persist_directory="../chroma_db",
    collection_name="documentation"
)

doc_urls = [
    "https://docs.python.org/3/tutorial/index.html",
    "https://fastapi.tiangolo.com/tutorial/",
    "https://fastapi.tiangolo.com/",
    "https://react.dev/",
    "https://platform.openai.com/docs/",
    "https://docs.anthropic.com/",
    "https://docs.aws.amazon.com/",
    "https://kubernetes.io/docs/",

]

# Scrape and add documents
for url in doc_urls:
    docs = vector_store.scrape_documentation(url, max_pages=20)
    vector_store.add_documents(docs)
```
